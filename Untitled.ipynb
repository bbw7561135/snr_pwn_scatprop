```python
def e_d(r):
    C = 2.938e+55
    return ((1.7283795e+57/(r**3))/(C))
```

```python
r_1 = 1.
r_2 = 2.
r_3 = 3.

x_1 = np.arange(0.1,r_1,0.001)
x_2 = np.arange(1,r_2,0.001)
x_3 = np.arange(2,r_3,0.001)

e_d_1 = e_d(r_1)
e_d_2 = e_d(r_2)
e_d_3 = e_d(r_3)

fig, ax = plt.subplots()
ax.plot(r_1, e_d_1)
ax.plot(r_2, e_d_2)
ax.plot(r_3, e_d_3)
ax.set_xlabel("Radius (parsecs)")
ax.set_ylabel("Electron Density (cm^3)")
plt.show()
```

```python
def e_d(r):
    M = 1.6E+31
    m = 1.7E-27
    mu = 1.3
    return ((3.*M)/(4.*np.pi*(r**3.)*mu*m))
```

```python
t_c1 = 250.
t_c2 = 1250.
t_c3 = 11500.
t_c4 = 511500.

def r(t):
    return (12.9)*((t/10**4)**(2/5))*((.75*10**51)/(8.814E+58))**(1/5)

x_1 = np.arange(10,t_c1,0.01)
x_2 = np.arange(250,t_c2,0.01)
x_3 = np.arange(1250,t_c3,0.01)
x_4 = np.arange(11500,t_c4,0.01)

r_1 = r(x_1)
r_2 = r(x_2)
r_3 = r(x_3)
r_4 = r(x_4)

e_d_1 = e_d(r_1)
e_d_2 = e_d(r_2)
e_d_3 = e_d(r_3)
e_d_4 = e_d(r_4)

fig, ax = plt.subplots()
ax.plot(x_1, e_d_1)
ax.plot(x_2, e_d_2)
ax.plot(x_3, e_d_3)
ax.plot(x_4, e_d_4)
ax.set_xlabel("Time (years)")
ax.set_ylabel("Electron Density")
plt.show()
```
